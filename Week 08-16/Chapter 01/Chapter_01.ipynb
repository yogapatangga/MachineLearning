## Sistem Pembelajaran Mesin (Machine Learning Systems)

Machine Learning adalah bidang yang memungkinkan mesin belajar dari data **tanpa diprogram secara eksplisit**. Sistem Machine Learning dapat diklasifikasikan menjadi beberapa kategori:

---

## 1. Berdasarkan Supervisi

| Tipe                   | Deskripsi                                                                 | Contoh Aplikasi                                      |
|------------------------|---------------------------------------------------------------------------|------------------------------------------------------|
| **Supervised Learning** | Menggunakan dataset berlabel untuk melatih model.                         | Klasifikasi (spam filter), regresi (harga mobil)     |
| **Unsupervised Learning** | Menggunakan data tanpa label, mencari pola sendiri.                       | Clustering, deteksi anomali, reduksi dimensi         |
| **Semisupervised Learning** | Menggabungkan data berlabel dan tidak berlabel.                        | Pengenalan wajah di layanan foto                     |
| **Reinforcement Learning** | Agen belajar dari penghargaan dan hukuman untuk menentukan kebijakan. | Robot belajar berjalan, AlphaGo                      |

---

## 2. Berdasarkan Pembelajaran Inkremental

| Tipe            | Deskripsi                                                                 |
|------------------|---------------------------------------------------------------------------|
| **Batch Learning** | Model dilatih sekaligus dengan semua data, tidak dapat diperbarui bertahap. |
| **Online Learning** | Model belajar secara bertahap dari data yang masuk terus-menerus.        |

---

## 3. Berdasarkan Metode Generalisasi

| Tipe                    | Deskripsi                                                                 |
|--------------------------|---------------------------------------------------------------------------|
| **Instance-based Learning** | Menghafal contoh, membandingkan kasus baru dengan ukuran kesamaan.     |
| **Model-based Learning**    | Membangun model dari data untuk membuat prediksi.                       |

---

## Tantangan Utama Machine Learning

- **Data Pelatihan yang Tidak Mencukupi**  
  → Algoritma butuh banyak data agar bekerja optimal.

- **Data Pelatihan Tidak Representatif**  
  → Model hanya akan bekerja baik jika data mencerminkan kasus nyata.

- **Data Berkualitas Buruk**  
  → Error, outlier, dan noise dapat mengganggu pembelajaran model.

- **Fitur yang Tidak Relevan**  
  → Pemilihan fitur sangat memengaruhi performa model.

- **Overfitting**  
  → Model terlalu kompleks, menangkap noise dan gagal generalisasi.

- **Underfitting**  
  → Model terlalu sederhana untuk mengenali pola dalam data.


## Reproduksi Kode
Berikut adalah reproduksi kode untuk model regresi linear yang memprediksi tingkat kepuasan hidup berdasarkan GDP per kapita:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load data
data = pd.read_csv("gdp-vs-happiness.csv")
```

```python
print(data.columns)
```

```python
print(data.to_string())
```

```python
X = data[["GDP per capita, PPP (constant 2021 international $)"]]
y = data[["Cantril ladder score"]]
```

```python
# Visualisasi data
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue")
plt.xlabel("GDP per capita (PPP, constant 2021 $)")
plt.ylabel("Life satisfaction (Cantril ladder score)")
plt.title("GDP per Kapita vs Kepuasan Hidup")
plt.grid(True)
plt.show()
```

```python
# Training model regresi linear
model = LinearRegression()
model.fit(X, y)
```

```python
# Prediksi untuk GDP per kapita baru, misal 40000
gdp_new = [[40000]]
prediksi = model.predict(gdp_new)
print(f"Prediksi kepuasan hidup untuk GDP per kapita 40,000: {prediksi[0][0]:.2f}")
```

```python
# Visualisasi hasil regresi linear
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data asli")
plt.plot(X, model.predict(X), color="red", linewidth=2, label="Regresi linear")
plt.xlabel("GDP per capita (PPP, constant 2021 $)")
plt.ylabel("Life satisfaction (Cantril ladder score)")
plt.title("Model Regresi Linear")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ambil data yang diperlukan
X = data["GDP per capita, PPP (constant 2021 international $)"]
y = data["Cantril ladder score"]
regions = data["World regions according to OWID"]
countries = data["Entity"]

# Warna untuk tiap region
colors = {
    'Africa': 'purple',
    'Asia': 'teal',
    'Europe': 'blue',
    'North America': 'coral',
    'Oceania': 'cyan',
    'South America': 'brown'
}

plt.figure(figsize=(12,7))
```

```python
# Plot per region dengan warna berbeda
for region, color in colors.items():
    subset = data[regions == region]
    plt.scatter(subset["GDP per capita, PPP (constant 2021 international $)"],
                subset["Cantril ladder score"],
                label=region, color=color, alpha=0.7, edgecolors='w', s=80)

plt.xlabel("GDP per capita (PPP, constant 2021 $)")
plt.ylabel("Life satisfaction (Cantril ladder score)")
```

```python
plt.figure(figsize=(12,7))

plt.xscale('log')  # skala log dulu sebelum plot

for region, color in colors.items():
    subset = data[regions == region]
    plt.scatter(subset["GDP per capita, PPP (constant 2021 international $)"],
                subset["Cantril ladder score"],
                label=region, color=color, alpha=0.7, edgecolors='w', s=80)

for i, country in enumerate(countries):
    gdp = X.iloc[i]
    life = y.iloc[i]
    if gdp > 50000 or life > 6.5 or country in ["Costa Rica", "Guatemala", "Bolivia", "Bangladesh", "Lebanon", "Botswana"]:
        plt.text(gdp * 1.02, life + 0.05, country, fontsize=8)

plt.xlabel("GDP per capita (PPP, constant 2021 $)")
plt.ylabel("Life satisfaction (Cantril ladder score)")
plt.title("GDP per Kapita vs Kepuasan Hidup")
plt.legend(title="Region")
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

# Penjelasan Kode Machine Learning  

## 1. Model Regresi Linear

Model pertama menggunakan pendekatan **model-based learning** dengan algoritma **Regresi Linear**.

### Tahapan:
- **Persiapan Data**  
  - Membaca data dari dua sumber: OECD (indeks kepuasan hidup) dan IMF (GDP per kapita).  
  - Menggabungkan kedua sumber menjadi satu dataset terpadu.
  
- **Visualisasi Data**  
  - Membuat *scatter plot* untuk menggambarkan hubungan antara GDP per kapita dan kepuasan hidup.

- **Pelatihan Model**  
  - Menggunakan `LinearRegression` untuk melatih model terhadap data.
  
- **Prediksi**  
  - Memprediksi kepuasan hidup negara **Cyprus** berdasarkan GDP per kapitanya ($22,587).

### Penjelasan Regresi Linear  
Model berusaha menemukan garis lurus terbaik dalam bentuk:
  
$$
y = \theta_0 + \theta_1 x
$$

- Di mana:
  - \( \theta_0 \): *Intercept* (nilai y saat x = 0)
  - \( \theta_1 \): *Slope* atau kemiringan garis
- Tujuannya adalah meminimalkan *mean squared error* (MSE) antara prediksi dan data asli.

---

## 2. Model k-Nearest Neighbors (k-NN)

Model kedua menggunakan pendekatan **instance-based learning** dengan algoritma **k-Nearest Neighbors Regressor**.

### Tahapan:
- **Persiapan dan Visualisasi Data**  
  - Sama seperti pada model regresi linear.

- **Pelatihan Model**  
  - Menggunakan `KNeighborsRegressor` dengan `n_neighbors=3`, artinya prediksi didasarkan pada 3 negara terdekat dari segi GDP per kapita.

- **Prediksi**  
  - Model mencari 3 negara dengan GDP per kapita paling mirip Cyprus, lalu mengambil **rata-rata** nilai kepuasan hidup mereka.

- **Analisis Tambahan**  
  - Menampilkan ketiga negara yang digunakan untuk prediksi sebagai pembanding.

### Penjelasan k-NN  
Berbeda dari regresi linear:
- k-NN **tidak membentuk model eksplisit**, melainkan menyimpan data pelatihan.
- Prediksi dilakukan berdasarkan **kemiripan** dengan instance yang telah ada.

---

## Hasil & Insight

Beberapa hal penting yang dapat disimpulkan dari kedua model:

### Perbandingan Pendekatan
| Aspek | Regresi Linear | k-NN |
|-------|----------------|------|
| Tipe Model | Eksplisit (model-based) | Implicit (instance-based) |
| Interpretasi | Sederhana, mudah dipahami | Lebih kompleks |
| Fleksibilitas | Kurang fleksibel untuk data non-linear | Lebih fleksibel terhadap pola lokal |
| Overfitting | Risiko lebih kecil | Lebih rentan jika k terlalu kecil |

### Insight Tambahan:
- **Pendekatan Berbeda, Hasil Serupa**  
  Keduanya bisa memberikan estimasi cukup baik, meskipun metodenya berbeda.
  
- **Pola Data Nyata Tidak Selalu Linear**  
  Negara dengan GDP tinggi tidak selalu memiliki skor kebahagiaan tertinggi.

- **Overfitting & Regularisasi**  
  Model kompleks seperti k-NN lebih berisiko overfit jika tidak divalidasi dengan baik.

- **Pentingnya Representasi Data**  
  Model akan berkinerja lebih baik jika data pelatihan mencerminkan data dunia nyata yang ingin diprediksi.

---

## Kesimpulan

Tantangan utama dalam machine learning adalah menyeimbangkan antara:
- **Sederhana** → agar tidak overfit,
- **Kompleks** → agar bisa menangkap pola yang bermakna.

Oleh karena itu, pemilihan fitur yang tepat, evaluasi menyeluruh, dan validasi silang adalah **kunci untuk generalisasi yang baik** terhadap data baru.
